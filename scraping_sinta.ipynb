```python
import selenium
import time
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import StaleElementReferenceException 
from selenium import webdriver

from tqdm import tqdm # beautify loop
import pandas as pd # dataframe things
```

```python
#Inisiasi 
s=Service('C:/SAINS DATA/chromedriver.exe')
options = webdriver.ChromeOptions()
options.add_experimental_option("detach", True)
options.add_argument('disable_infobars')

driver = webdriver.Chrome(service=s, options=options)
url="https://sinta.kemdikbud.go.id/authors"

driver.get(url)
# driver.maximize_window()
```

```python
df = pd.read_excel('Data Author-FULL.xlsx')
df
```

```python
# df.iloc[df['Nama'] == "A IND'AH CAMELIA", :]
# df.loc[df['Nama'] == "ZULFAYANDI PAWANIS", :]
# 70764
# 72796

# the name that we use for keyword
data_nama = list(df.iloc[70764:72797, 0])
data_nama[:5]
```

```python
sinta_dict = {'Nama' : [], 
              'Universitas' : [],
              'Department' : [],
              'SINTA Score 3Yr' : [],
              'SINTA Score Overall' : [], 
              'Subject 1' : [], 
              'Subject 2' : [], 
              'Subject 3' : [], 
              'Subject 4' : [], 
              'Subject 5' : [], 
              'Index 1' : [], 
              'Index 2' : [], 
              'Index 3' : []
              }
```

## Scraping

```python
for i in tqdm(data_nama):    # iterate through all names
    search = driver.find_element(By.NAME, 'q')      # locate search box
    search.clear()      # clear searchbox
    search.send_keys(i) # input names to searchbox
    search.send_keys(Keys.ENTER) # enter
    time.sleep(0.2) # freeze for a moment

    parent = driver.find_elements(By.CSS_SELECTOR, 'div[class = "list-item row mb-4"]') # locate element parent
    for subparent in parent:
        campus = subparent.find_element(By.CLASS_NAME, 'profile-affil').text # locate university
        if campus == 'Universitas Airlangga': # filter: unair
            sinta_dict['Universitas'].append(campus) 
            
            name = subparent.find_element(By.CLASS_NAME, 'profile-name').text # locate name
            sinta_dict['Nama'].append(name)
            
            prodi = subparent.find_element(By.CSS_SELECTOR, 'div[class = "profile-dept"]').text # locate department
            sinta_dict['Department'].append(prodi)
            
            sinta_score = subparent.find_elements(By.CLASS_NAME, 'pr-num') # locate scores
            sinta_dict['SINTA Score 3Yr'].append(sinta_score[0].text)
            sinta_dict['SINTA Score Overall'].append(sinta_score[-1].text)
            
            subjects = subparent.find_elements(By.TAG_NAME, 'li') # locate subjects
            for i in range(5):
                try:
                    sinta_dict[f'Subject {i+1}'].append(subjects[i].text)
                except IndexError: 
                    sinta_dict[f'Subject {i+1}'].append(None) #if the subject <= 5, fill the list with none
            
            table = subparent.find_elements(By.TAG_NAME, 'td') # locate index
            table_only_int = table[2::3]
            for i in range(3):
                sinta_dict[f'Index {i+1}'].append(int(table_only_int[i].text)) 
```

```python
# as dataframe
sinta_df = pd.DataFrame(sinta_dict)
sinta_df
```

```python
sinta_df.drop_duplicates()
```

```python
sinta_df.to_excel('sinta_scraped_no_duplicate.xlsx')
print('done!')
```
