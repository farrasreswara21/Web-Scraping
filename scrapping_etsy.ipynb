```python
import pandas as pd

import selenium
import time
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException 
from selenium import webdriver
```

```python
#Inisiasi 
s=Service('C:/SAINS DATA/chromedriver.exe')
options = webdriver.ChromeOptions()
options.add_experimental_option("detach", True)
options.add_argument('disable_infobars')
# options.headless = True

driver = webdriver.Chrome(service=s, options=options)
url="https://www.etsy.com/search?q=handmade%20jewelry&ref=auto-1&as_prefix=hand"
url1 = 'https://www.etsy.com/listing/639151531/ready-to-ship-crystal-necklace-for-women?ga_order=most_relevant&ga_search_type=all&ga_view_type=gallery&ga_search_query=handmade+jewelry&ref=sc_gallery-1-1&pro=1&sts=1&plkey=03dde4a8e3389d4a03632cf487cea99876ada361%3A639151531'
driver.get(url)
```

```python
# get links
store_links = []

links_parent = driver.find_element(By.CSS_SELECTOR, 'ul[class = "wt-grid wt-grid--block wt-pl-xs-0 tab-reorder-container"]')
links = links_parent.find_elements(By.TAG_NAME, 'a')
for i in links:
    if i.get_attribute('title') != 'More like this' : 
        # store_links.append(i.get_attribute('href'))
        print(i.get_attribute('href'))
```

```python
# next 
next = driver.find_elements(By.CSS_SELECTOR, 'li[class = "wt-action-group__item-container"]')
next[-1].click()
```

```python
store_links = []
while True: 
    if len(store_links) < 100:
        print(len(store_links))
        links_parent = driver.find_element(By.CSS_SELECTOR, 'ul[class = "wt-grid wt-grid--block wt-pl-xs-0 tab-reorder-container"]')
        links = links_parent.find_elements(By.TAG_NAME, 'a')
        for i in links:
            if i.get_attribute('title') != 'More like this' : 
                store_links.append(i.get_attribute('href'))
        
        next = driver.find_elements(By.CSS_SELECTOR, 'li[class = "wt-action-group__item-container"]')
        next[-1].click()
        
        time.sleep(2.5)
    else: 
        break
```

```python
# get price
price = driver.find_element(By.CSS_SELECTOR, 'div[data-buy-box-region = "price"]')
price_child = price.find_element(By.CSS_SELECTOR, 'span[class = ""]')
print(price_child.text)
```

```python
#get store name
store_parent = driver.find_element(By.CSS_SELECTOR, 'div[data-action = "follow-shop-listing-header"]')
store = store_parent.find_element(By.CSS_SELECTOR, 'a[class = "wt-text-link-no-underline"]')
print(store.text)

# store = store_parent.find_elements(By.CSS_SELECTOR, '*')
# print([i.text for i in store])
```

```python
# get productname 
product = driver.find_element(By.CSS_SELECTOR, 'h1[class = "wt-text-body-03 wt-line-height-tight wt-break-word"]')
print(product.text)
```

```python
# get total sales 
total_sales = store_parent.find_element(By.CSS_SELECTOR, 'span[class = "wt-text-caption"]')
print(total_sales.text)
```

```python
# get cost to ship 
ship_cost_parent = driver.find_element(By.CSS_SELECTOR, 'div[data-estimated-shipping]')
ship_cost = ship_cost_parent.find_element(By.CSS_SELECTOR, 'p[class = "wt-text-body-03 wt-mt-xs-1 wt-line-height-tight"]')
print(ship_cost.text)

# ship_cost = ship_cost_parent.find_elements(By.CSS_SELECTOR, '*')
# # print([i.text for i in ship_cost])
# print(ship_cost[2].tag_name)
```

```python
# get star rating 
star_parent = driver.find_element(By.ID, 'reviews')

star_in_num = star_parent.find_element(By.CSS_SELECTOR, 'input[name = "rating"]')
print(star_in_num.get_attribute('value'))

star = star_parent.find_element(By.CSS_SELECTOR, 'span[class = "wt-screen-reader-only"]')
print(star.text)
```

```python
list_price = []
list_storename = []
list_productname = []
list_totalsales = []
list_costtoship = []
list_starnum = []
list_star = []
n = 1 

for ling in store_links : 
    driver.get(ling)
    print(ling)
    time.sleep(0.5)
    
    # GET PRICE
    price = driver.find_element(By.CSS_SELECTOR, 'div[data-buy-box-region = "price"]')
    try :     
        price_child = price.find_element(By.CSS_SELECTOR, 'span[class = ""]')
        list_price.append(price_child.text)
    except NoSuchElementException : 
        price_child2 = price.find_element(By.CSS_SELECTOR, 'p[class = "wt-text-title-03 wt-mr-xs-2"]')
        list_price.append(price_child2.text)
        
    #GET STORE NAME
    store_parent = driver.find_element(By.CSS_SELECTOR, 'div[data-action = "follow-shop-listing-header"]')
    store = store_parent.find_element(By.CSS_SELECTOR, 'a[class = "wt-text-link-no-underline"]')
    list_storename.append(store.text)
    
    # GET PRODUCTNAME 
    product = driver.find_element(By.CSS_SELECTOR, 'h1[class = "wt-text-body-03 wt-line-height-tight wt-break-word"]')
    list_productname.append(product.text)
    
    # GET TOTAL SALES 
    try : 
        total_sales = store_parent.find_element(By.CSS_SELECTOR, 'span[class = "wt-text-caption"]')
        list_totalsales.append(total_sales.text)
    except NoSuchElementException : 
        list_totalsales.append(0)
    
    # GET COST TO SHIP 
    ship_cost_parent = driver.find_element(By.CSS_SELECTOR, 'div[data-estimated-shipping]')
    ship_cost = ship_cost_parent.find_element(By.CSS_SELECTOR, 'p[class = "wt-text-body-03 wt-mt-xs-1 wt-line-height-tight"]')
    list_costtoship.append(ship_cost.text)
    
    # GET STAR RATING (NUMBER)
    star_parent = driver.find_element(By.ID, 'reviews')
    star_in_num = star_parent.find_element(By.CSS_SELECTOR, 'input[name = "rating"]')
    list_starnum.append(star_in_num.get_attribute('value'))

    #GET STAR RATING (STAR)
    star = star_parent.find_element(By.CSS_SELECTOR, 'span[class = "wt-screen-reader-only"]')
    list_star.append(star.text)
    
    
    print(f'{n}. {product.text}')
    n += 1
```

```python
dict_etsy = {
    'Product Name' : list_productname,
    'Store' : list_storename,
    'Price' : list_price, 
    'Total Sales' : list_totalsales,  
    'Rating' : list_starnum, 
    'Star' : list_star, 
    'Cost to ship' : list_costtoship
}

# Make Dataframe
df_etsy = pd.DataFrame(dict_etsy)
df_etsy.head(10)
```
